```python
import os
import numpy as np
import pandas as pd
```

```python
results = []
results_dir = '/home/thompsjame1/grid-output/4676488_rl_continuous_control_baseline_experiements/training_evaluations'

for filename in os.listdir(results_dir):
    if not filename.endswith('.npz'):
        continue

    model = filename.split('_')[0]
    env = filename.split('_')[1]
    steps = filename.split('_')[2]
    seed = filename.split('_')[3].split('.')[0]
    npz = np.load(os.path.join(results_dir, filename))

    results.append({
        'model': model,
        'env': env,
        'total_steps': steps,
        'seed': seed,
        'timesteps': npz['timesteps'],
        'rewards': npz['results'],
        'ep_lengths': npz['ep_lengths'],
    })

df = pd.DataFrame(results)

df.explode(['timesteps', 'rewards', 'ep_lengths'])
```
